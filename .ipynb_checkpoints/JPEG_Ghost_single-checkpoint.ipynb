```python
plt.rcParams["font.size"]=10
plt.rcParams["figure.figsize"]=(4.0, 2.0)
plt.rcParams["figure.dpi"]= 300



# 基準となる画像を読み込む
path = '/Users/yoshihisafurushita/Desktop/JPEG_ghost_single/1_1stQF20.jpg'
reference_image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

# SSDの結果を格納するリストを作成
ssd_results = []

# 複数の画像を処理するループ
for i in range(5, 26):  # 例として5つの画像を処理
    # 画像を読み込む
    image = cv2.imread(f"/Users/yoshihisafurushita/Desktop/JPEG_ghost_single/1_1stQF20_{i}.jpg", cv2.IMREAD_GRAYSCALE)
    
    # 画像サイズの一致を確認し、リサイズする
    if reference_image.shape != image.shape:
        image = cv2.resize(image, (reference_image.shape[1], reference_image.shape[0]))
    
    # SSDを計算して結果をリストに追加
    ssd = np.sum((reference_image.astype(float) - image.astype(float)) ** 2)
    ssd_results.append(ssd)

# SSDの結果をグラフにプロット
plt.figure()

plt.ylim(0, 3.5e8)
# plt.ylim(min(ssd_results), max(ssd_results) * 1.１)

plt.plot(range(5, 26), ssd_results, marker='.', linestyle='-')
plt.xlabel('Quality Factor')
plt.ylabel('SSD Value')
# plt.grid(True)

plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))

highlight_index = 15  # x軸が20の場合、インデックス15になります
plt.plot(20, ssd_results[highlight_index], marker='.', markersize=8, color='red')
plt.savefig('/Users/yoshihisafurushita/Desktop/coding_ghost_single.pdf', bbox_inches='tight', pad_inches=0.0)

# plt.show()
```
